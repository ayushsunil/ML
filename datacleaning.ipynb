```python
import pandas as pd
dc=pd.read_csv("https://raw.githubusercontent.com/Jcharis/Data-Cleaning-Practical-Examples/master/unclean_data1.csv")
print(dc)
```

```python
dc.head(14)
```

```python
dc.isnull().sum()
```

```python
nmc=dc.select_dtypes(include='number').columns
dc[nmc]=dc[nmc].fillna(dc[nmc].mean())
nonmc=dc.select_dtypes(exclude='number').columns
dc[nonmc]=dc[nonmc].fillna(dc[nonmc].mode().iloc[0])
```

```python
dc.drop_duplicates(inplace=True)
```

```python
dc.to_csv("clean_data.csv",index=False)
print(dc)
dc.head(14)
```
